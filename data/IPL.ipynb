```python
# Team w max wins
team_names=team["Team_Name"]
team_ids=matches['Match_Winner_Id'].value_counts()
team_ids
win_count=[]
for key,value in team_ids.items():
    val={
        "team_name": team_names[key-1],
        "count": value
    }
    win_count.append(val)
    
print(win_count)
# json_win = json.dumps(win_count)
# text_file = open("team_wins.json", "w")
# text_file.write(json_win)
# text_file.close()
```

```python
#Players with max motm
player_names=player["Player_Name"]
motm_ids=matches["Man_Of_The_Match_Id"].value_counts()
motm_count=[]
for key,value in motm_ids.items():
    val={
        "player_name": player_names[key-1],
        "count": value
    }
    motm_count.append(val)

print(motm_count[:10])
# json_win = json.dumps(motm_count[:10])
# text_file = open("motm.json", "w")
# text_file.write(json_win)
# text_file.close()
```

```python
#teams with max toss wins
team_ids=matches["Toss_Winner_Id"].value_counts()
toss_count=[]
for key,value in team_ids.items():
    val={
        "team_name": team_names[key-1],
        "count": value
    }
    toss_count.append(val)

print(toss_count)
# json_win = json.dumps(toss_count)
# text_file = open("toss_win.json", "w")
# text_file.write(json_win)
# text_file.close()
```

```python
#win types
team_ids=matches.groupby('Win_Type')['Match_Id'].count()
res_count=[]
for key,value in team_ids.items():
    val={
        "win_type": key,
        "count": value
    }
    res_count.append(val)

print(res_count)
# json_win = json.dumps(res_count)
# text_file = open("win_type.json", "w")
# text_file.write(json_win)
# text_file.close()
```

```python
# Team with max boundaries
team_bounds=[]
for i in range(1,14):
    team_data=ballbyball[ballbyball['Team_Batting_Id']==i]
    td=team_data[(team_data['Batsman_Scored']!=None)]['Batsman_Scored']
    count=0
    for key, value in td.items():
        if value>="3":
            count+=1
    val={
        "team_name": team_names[i-1],
        "count": count
    }
    team_bounds.append(val)
    
team_bounds=sorted(team_bounds, key = lambda i: i['count'], reverse=True) 
print(team_bounds)
# json_win = json.dumps(team_bounds)
# text_file = open("team_bounds.json", "w")
# text_file.write(json_win)
# text_file.close()
```

```python
# most wins each season
season_wins=[]
for i in range(1,10):
    season=matches[matches['Season_Id']==i]
    idx=season['Match_Winner_Id'].value_counts().idxmax
    count=season['Match_Winner_Id'].value_counts()[idx]
    val={
        "season": int(i),
        "team": team_names[idx],
        "count": int(count)
    }
    season_wins.append(val)
print(season_wins)
# json_win = json.dumps(season_wins)
# text_file = open("season_wins.json", "w")
# text_file.write(json_win)
# text_file.close()
```

```python
team_score=[]
for i in range(1,14):
    team_data=ballbyball[ballbyball['Team_Batting_Id']==i]
    num_matches=team_data['Match_Id'].nunique()
    td=team_data[(team_data['Batsman_Scored']!=None)]['Batsman_Scored']
    count, balls, mx, mn, mc=0,0,0,500,0
    for key, value in td.items():
        if balls==120:
            mx=max(mx,mc)
            mn=min(mn,mc)
            count+=mc
            balls, mc = 0,0
        elif value!= "Do_nothing":
            try:
                mc+=int(value)
                balls+=1
            except:
                pass
    val={
            "team_name": team_names[i-1],
            "avg": count/num_matches,
            "max": mx,
            "min": mn
        }
    team_score.append(val)
    
# team_score=sorted(team_score, key = lambda i: i['count'], reverse=True) 
print(team_score)
json_win = json.dumps(team_score)
text_file = open("team_scores.json", "w")
text_file.write(json_win)
text_file.close()
```
